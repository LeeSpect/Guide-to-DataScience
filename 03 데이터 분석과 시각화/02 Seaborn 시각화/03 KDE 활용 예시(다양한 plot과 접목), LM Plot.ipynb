```python
import pandas as pd
import seaborn as sns

body_df = pd.read_csv('Desktop/body.csv', index_col=0)
body_df.head()
```

```python
# 히스토그램
body_df.plot(kind='hist', y='Height', bins=15)
```

```python
# Seaborn을 사용하여 히스토그램 위에 KDE Plot을 출력할 수 있음
sns.distplot(body_df['Height'], bins=15)  # distplot 사용
```

```python
# Box Plot(박스 플롯)
body_df.plot(kind='box', y='Height')
```

```python
sns.violinplot(y=body_df['Height'])  # violinplot 사용  # KDE 그래프를 90도 회전해서 대칭시킨 형태
```

```python
# Scatter Plot 사용, 키와 몸무게의 연관성 보기
body_df.plot(kind='scatter', x='Height', y='Weight')
```

```python
sns.kdeplot(body_df['Height'], body_df['Weight'])  # 등고선 표현   # 키와 몸무게의 연관성과 분포를 동시에 볼 수 있음
```

```python
sns.kdeplot(body_df['Height'])
```

```python
sns.kdeplot(body_df['Weight'])
```

```python
sns.lmplot(data=body_df, x='Height', y='Weight')
```
