```python
import pandas as pd
import seaborn as sns

laptops_df = pd.read_csv('Desktop/laptops.csv')
laptops_df.head()
```

```python
# 운영체제의 column 출력
laptops_df['os']  # Pandas Series형으로 출력
```

```python
laptops_df['os'].unique()  # 어떤 값들이 있는지 추려서 array형으로 출력
```

```python
# 각 OS가 어떤 특성을 가지고 있는지 알고 싶다(ex. 어떤 OS가 비싼지)
# 요소들을 카테고리 별로 비교하고 싶으면 다음과 같이 하면 됨.
sns.catplot(data=laptops_df, x='os', y='price', kind='box')
```

```python
sns.catplot(data=laptops_df, x='os', y='price', kind='violin')
```

```python
sns.catplot(data=laptops_df, x='os', y='price', kind='strip')  # strip을 사용하면 모든 값들을 출력함
```

```python
laptops_df['processor_brand'].unique()  # 위 그래프의 값들 중 intel, amd를 쓰는 것은 어떤 것인지 알고 싶음.
```

```python
sns.catplot(data=laptops_df, x='os', y='price', kind='strip', hue='processor_brand')
# hue는 색을 의미. 즉 여기에서는 processor_brand에 따라 색을 다르게 하라는 뜻.
```

```python
# 위 그래프에서 점들이 많이 몰려있는 하단 부분은 보기가 불편함. 이를 해결하기 위해선 kind를 swarm으로 변경
sns.catplot(data=laptops_df, x='os', y='price', kind='swarm', hue='processor_brand')
```
