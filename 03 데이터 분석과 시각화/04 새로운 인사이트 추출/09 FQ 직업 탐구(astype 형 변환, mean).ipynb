```python
import pandas as pd

df = pd.read_csv('Desktop/occupations.csv')
df
```

```python
female = df['gender'] == 'F'

df['Gender'] = None
df.loc[female, 'Gender'] = 1
df
```

```python
df_groups = df.groupby('occupation')
(df_groups.count()['Gender'] / df_groups.count()['gender']).sort_values(ascending=False)
```

```python
# 모범답안
df = pd.read_csv('Desktop/occupations.csv')

occupation_group = df.groupby('occupation')
occupation_group
```

```python
df.loc[df['gender'] == 'M', 'gender'] = 0
df.loc[df['gender'] == 'F', 'gender'] = 1

df['gender'] = df['gender'].astype(int)  # astype 메소드로 type 변경
occupation_group.mean()['gender'].sort_values(ascending=False)
```
