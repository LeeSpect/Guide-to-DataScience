```python
# Seaborn(Statistical Data Visualization) 통계를 기반으로 한 데이터 시각화

# PDF = Probability Density Function = 확률 밀도 함수
# PDF(확률 밀도 함수)는 데이터셋의 분포를 나타낸다.
# 특정 구간의 확률은 그래프 아래 그 구간의 면적과 동일하다.
# 그래프 아래의 모든 면적을 더하면 1이 된다.

#KDE(Kernel Density Estimation)
```

```python
import pandas as pd
import seaborn as sns

body_df = pd.read_csv('Desktop/body.csv', index_col=0)
body_df.head()
```

```python
# Height만 다루기
body_df['Height']  # Pandas Series 형식으로 출력됨
```

```python
# 각 키가 몇번씩 나오는지 출력
body_df['Height'].value_counts()
```

```python
# 출력된 내용을 키 순으로 출력(키 순 = index 순)
body_df['Height'].value_counts().sort_index()
```

```python
# 위 내용을 그래프로 출력
body_df['Height'].value_counts().sort_index().plot()
```

```python
# 위 그래프를 매끄러운 곡선 그래프로 출력 ** Seaborn 사용 **
sns.kdeplot(body_df['Height'])  # PDF로 출력
```

```python
# 위 그래프는 정확한 그래프가 아니라 추측을 통해 만들어진 그래프임
# 추측을 어느정도 감도로 할지 정하려면 bw라는 파라미터를 조정
sns.kdeplot(body_df['Height'], bw=0.05)
```

```python
sns.kdeplot(body_df['Height'], bw=0.5)  # bw를 크게 할수록 그래프가 부드러워짐
```
