```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# 빈 리스트 생성
rating_pages = []

for year in range(2010, 2019):
    for month in range(1, 13):
        for week in range(0, 5):
            # HTML 코드 받아오기
            response = requests.get("https://workey.codeit.kr/ratings/index?year= "+ str(year) + '&month=' + str(month) + '&weekIndex=' + str(week))
            
            # BeautifulSoup 타입으로 변환하기
            soup = BeautifulSoup(response.text, 'html.parser')

            # ".csrch_tip" 클래스가 없을 때만 HTML 코드를 리스트에 담기
            if len(soup.select('.row')) > 1:
                period = soup.select('div.date-pick > div.search-box > section > select#weekSelectBox > option')
                rank = soup.select('tr.row > td.rank')
                channel = soup.select('tr.row > td.channel')
                program = soup.select('tr.row > td.program')
                rating = soup.select('tr.row > td.percent')
                
                for i in range(len(period)):  # len(period)라고 하면 안 됨
                    record = []
                    record.append(period[0].text)
                    record.append(rank[i].text)
                    record.append(channel[i].text)
                    record.append(program[i].text)
                    record.append(rating[i].text)
                    rating_pages.append(record)
                
            else:
                break

df = pd.DataFrame(data = rating_pages, columns = ['period', 'rank', 'channel', 'program', 'rating'])
                
# 결과 출력
df.head()
```
